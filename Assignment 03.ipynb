```python
import pandas as pd
import numpy as nm
```

```python
data= pd.read_csv("netflix_titles.csv")
data.head()
```

```python
def count_cast(val):
    if(pd.isnull(val)==True):
        return 0;
    else:
        arr=val.split()
        return len(arr)
    
    
    
data["Number of cast"]=data["cast"].apply(count_cast)
data.head()
```

```python
data1=data.groupby(["type","release_year"])
data1.first()
```

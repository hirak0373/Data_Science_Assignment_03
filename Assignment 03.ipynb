```python
import pandas as pd
import numpy as nm
```

```python
data= pd.read_csv("netflix_data.csv")
data.head()
```

```python
data["cast"].isnull().sum()
```

```python
def count_cast(val):
    if(pd.isnull(val)==True):
        return 0;
    else:
        arr=val.split()
        return len(arr)
    
    
    
data["Number of cast"]=data["cast"].apply(count_cast)
data.head()
```

```python
data1=data.groupby(["type","genre"])
data1.first()
```

```python
data2=data.groupby("type")["genre"].value_counts()
data2
```
